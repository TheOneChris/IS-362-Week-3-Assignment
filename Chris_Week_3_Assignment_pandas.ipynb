```python
import pandas as pan
import numpy as num
from matplotlib import pyplot as pyt

%matplotlib inline

travdist = pan.Series([55,120,179,227,263,315,364,399,445,489])
travdist.index += 1

daynum1 = pan.Series(travdist.iloc[0])
daynum1.index += 1

xdays = pan.Series(num.diff(travdist))
xdays.index += 2

numdays = [daynum1, xdays]
dailydist = pan.concat(numdays)

pyt.plot(dailydist, linestyle='--', marker='o')
pyt.ylabel('Miles traveled')
pyt.xlabel('Day number')
pyt.suptitle('Daily Miles Per Day')
pyt.grid()
pyt.show()

travdist.columns  = ['Day #', 'Traveled Distance']
print(travdist.columns)
print(travdist)
print('Total distanced traveled on bike = ',travdist.values[-1])
```
